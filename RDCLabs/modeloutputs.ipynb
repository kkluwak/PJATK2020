```python
import dataprocesing as dp

def wykresy_markerow(path):
    c = c3d(path)
    n_markers = ["LHipAngles","RHipAngles","LKneePower","RKneePower","RShoulderForce","LShoulderForce","LElbowMoment","RElbowMoment"]
    axes = ["x","y","z"]
    body = path.split('-')[3]+":"
    p,k = dp.read_labels(path,200)
    for mark in markers:
        n = c['parameters']['POINT']['LABELS']['value'][121:198].index(body+mark) 
        for i in range(3):
            for j in range(len(p)):
                plt.plot(c['data']['points'][i][n][p[j]:k[j]])
            plt.title("OÅ› "+axes[i]+"   "+n_markers[i])
            
            plt.show()
```
