```python
import dataprocesing as dp

def wykresy_markerow(path):
    c = c3d(path)
    n_markers = ['LGroundReactionForce', 'LNormalisedGRF', 'LAnkleForce', 'RAnkleForce', 'RKneeForce', 'LKneeForce', 'RHipForce', 'LHipForce', 'LWaistForce', 'RWaistForce', 'LNeckForce', 'RNeckForce', 'LShoulderForce', 'RShoulderForce', 'LElbowForce', 'RElbowForce', 'LWristForce', 'RWristForce']
    axes = ["x","y","z"]
    body = path.split('-')[3]+":"
    p,k = dp.read_labels(path,200)
    gg = -1
    for mark in n_markers:
        n = c['parameters']['POINT']['LABELS']['value'][:].index(body+mark)
        print(body+mark)
        gg = gg+1
        for i in range(3):
            
            for j in range(len(p)):
                plt.plot(c['data']['points'][i][n][p[j]:k[j]])
            plt.title("OÅ› "+axes[i]+"   "+n_markers[gg])
    
            plt.show()
```
