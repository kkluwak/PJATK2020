# Dostep do bazy danych miecza

```python
import h5py
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# f = h5py.File('Z:/gait_database/hma_basics_with_analog_preprocesing.h5', 'r')
f=h5py.File(r'\\172.16.10.4\Praktyki2020\gait_database\\kk_sword_basic.h5', 'r')
```

```python
print(list(f.keys()))
```

```python
x = f['x']
y = f['y_person']
```

```python
print('(samples, timesteps, features) (samples,label) -> ', x.shape, y.shape)
```

```python
fig, axs = plt.subplots(1,2)

y_df = pd.DataFrame(y)
y_df = y_df.rename(columns={0:'Labels'})
y_df_count = y_df['Labels'].value_counts()
y_df_count.plot.bar(ax=axs[0])
y_df_count.plot(kind='pie', figsize=(15, 6), autopct='%1.1f%%', shadow=True, labels=None, pctdistance=1.12, ax=axs[1])
plt.legend(labels=y_df_count.index, loc='upper left') 
plt.show()
```

```python
y = f['y_move']

fig, axs = plt.subplots(1,2)

y_df = pd.DataFrame(y)
y_df = y_df.rename(columns={0:'Labels'})
y_df_count = y_df['Labels'].value_counts()
y_df_count.plot.bar(ax=axs[0])
y_df_count.plot(kind='pie', figsize=(15, 6), autopct='%1.1f%%', shadow=True, labels=None, pctdistance=1.12, ax=axs[1])
plt.legend(labels=y_df_count.index, loc='upper left') 
plt.show()
```
